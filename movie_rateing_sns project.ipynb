```python
# python is not vectorize programming language
# Building dashboards (dashboard - combination of chats)

sns.set_style('darkgrid')

f, axes = plt.subplots(2, 2, figsize=(15, 15))

# 1. Budget vs Audience
k1 = sns.kdeplot(
    data=movies,
    x='BudgetMillions',
    y='AudienceRating',
    fill=False,
    cmap='Greens',
    ax=axes[0, 0]
)

# 2. Budget vs Critic
k2 = sns.kdeplot(
    data=movies,
    x='BudgetMillions',
    y='CriticRating',
    fill=False,
    cmap='Blues',
    ax=axes[0, 1]
)

# Set same x-limits
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# 3. Violin plot (Drama only)
z = sns.violinplot(
    data=movies[movies.Genre == 'Drama'],
    x='Year',
    y='CriticRating',
    ax=axes[1, 0]
)

# 4. Critic vs Audience KDE
k4 = sns.kdeplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    fill=True,
    cmap='Reds',
    levels=10,
    ax=axes[1, 1]
)

# Titles (important for dashboard clarity)
axes[0, 0].set_title("Budget vs Audience Rating")
axes[0, 1].set_title("Budget vs Critic Rating")
axes[1, 0].set_title("Drama: Critic Rating by Year")
axes[1, 1].set_title("Critic vs Audience Density")

plt.tight_layout()
plt.show()
```

```python
# How can you style your dashboard  using different color map

# python is not vectorize programming language
# Building dashboards (dashboard - combination of chats)


sns.set_style('dark', {'axes.facecolor': 'black'})

f, axes = plt.subplots(2, 2, figsize=(15, 15))

# [0,0] Budget vs Audience
k1 = sns.kdeplot(
    data=movies,
    x='BudgetMillions',
    y='AudienceRating',
    fill=False,
    cmap='inferno',
    ax=axes[0, 0]
)

# optional contour overlay
sns.kdeplot(
    data=movies,
    x='BudgetMillions',
    y='AudienceRating',
    cmap='cool',
    ax=axes[0, 0]
)

# [0,1] Budget vs Critic
k2 = sns.kdeplot(
    data=movies,
    x='BudgetMillions',
    y='CriticRating',
    fill=True,
    cmap='inferno',
    ax=axes[0, 1]
)

sns.kdeplot(
    data=movies,
    x='BudgetMillions',
    y='CriticRating',
    cmap='cool',
    ax=axes[0, 1]
)

# [1,0] Violin plot
z = sns.violinplot(
    data=movies[movies.Genre == 'Drama'],
    x='Year',
    y='CriticRating',
    palette='coolwarm',
    ax=axes[1, 0]
)

# [1,1] Critic vs Audience
k4 = sns.kdeplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    fill=True,
    cmap='Blues_r',
    ax=axes[1, 1]
)

sns.kdeplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    cmap='gist_gray_r',
    ax=axes[1, 1]
)

# axis limits
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# titles (important!)
axes[0, 0].set_title("Budget vs Audience")
axes[0, 1].set_title("Budget vs Critic")
axes[1, 0].set_title("Drama Critic Ratings by Year")
axes[1, 1].set_title("Critic vs Audience")

plt.tight_layout()
plt.show()
```
